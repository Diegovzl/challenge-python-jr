# Resolución del desafío.

```python
# 1) Generar un gráfico, un line chart donde el eje x sea la fecha y el eje y sea el active power para finalmente guardarlo.
# Graficar la generacion de energia diaria por planta.
for archivo in archivos:
  energy_per_day = pd.read_excel(archivo, usecols=["id_p", "fecha_im", "active_power_im", "active_energy_im"])

  #Eliminar posibles datos anormales.
  energy_per_day["active_power_im"] = pd.to_numeric(energy_per_day["active_power_im"], errors='coerce')

  energy_per_day["active_energy_im"] = pd.to_numeric(energy_per_day["active_energy_im"], errors='coerce')
  # Es necesario cambiar el formato de la fecha.
  energy_per_day['fecha_im'] = pd.to_datetime(energy_per_day['fecha_im'], format='%Y-%m-%d %H:%M:%S')

  #Creacion del grafico.
  with plt.style.context('ggplot'):
    fig, ax = plt.subplots(figsize = (16, 8), dpi=200)
    ax.title.set_text(f"Generación de energía diaria planta Id: {energy_per_day['id_p'][0]}")
    # plt.xticks(ticks=np.arange(energy_per_day['fecha_im'].min(), energy_per_day['fecha_im'].max()), 
    #           fontsize=14, 
    #           rotation=45)

    ax.set_xlabel('Hora', fontsize=18)
    ax.set_ylabel('Active power ', fontsize=16)

    ax.plot(list(energy_per_day['fecha_im']),
            list(energy_per_day['active_power_im']), 
            c='crimson')
            
    ax.xaxis.set_major_locator(md.MinuteLocator(byminute = [0,30]))
    ax.xaxis.set_major_formatter(md.DateFormatter('%H:%M'))
    plt.setp(ax.xaxis.get_majorticklabels(), rotation = 45)

  #Crear la ruta y nombre para el grafico a guardar.
  char_route = f'{char_Dir}/{archivo.split("/")[-1].split(".")[0]}.png' # Si se ejecuta el cuadreno de forma local, 
            
  #Usamos la funcion savefig de Matplotlib para almacenar el grafico creado.
  fig.savefig(char_route)
  plt. close()
  # 2) Salvar el la suma por día del active power, el valor máximo y mínimo del active energy y por último, 
  #la ruta donde se encuentra el grafico generado

  #Ruta para almacenar el archivo de texto generado.
  text_file_route = f'{text_Dir}/{archivo.split("/")[-1].split(".")[0]}.txt'# Si se ejecuta el cuadreno de forma local, 
                                                                        #es necesario reemplazar el slash (/) por doble Backslash (\\)

  #Obtener active power, el valor máximo y mínimo del active energy desde el marco de datos.
  lines = []
  lines.append(f"Id planta:{energy_per_day['id_p'][0]}")
  lines.append(f"Suma Diaria del active power:{energy_per_day['active_power_im'].sum()}")
  lines.append(f"Valor máximo del active energy:{energy_per_day['active_energy_im'].max()}")
  lines.append(f"valor mínimo del active energy:{energy_per_day['active_energy_im'].min()}")
  lines.append(f"{char_route}")

  # Abrir archivo en modo escritura.
  with open(text_file_route, 'w') as f:
    f.write('\n'.join(lines))
```

```python
# 3)Imprimir en consola la suma total del active power por día, de todas las plantas.

#Obtener las rutas de todos los archivos .txt almacenados en el directorio de archivos de texto.
files = glob(text_Dir + '/*.txt')

#Iterar sobre los archivos adquiridos.
for day in files:
  #Se abre el fichero en modo lectura para obtener la suma total de active power.
  myfile = open(day, "r")
  print(myfile.readline())
  print(myfile.readline())
```
